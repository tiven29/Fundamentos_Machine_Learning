```python
# Creación aleatoria de datos
import numpy as np
import random
random.seed(42)
```

```python
#Creao una variable de tipo lista vacia
nivel_estudios = []
#CReando o ingresando a mi lista la varaible Secuariada  y creo 110 veces secundarria
nivel_estudios.extend(['Secundaria']*110)
nivel_estudios.extend(['Bachillerato']*172)
nivel_estudios.extend(['Grado']*145)
nivel_estudios.extend(['Master']*52)
nivel_estudios.extend(['Doctorado']*21)

nivel_estudios = np.random.permutation(nivel_estudios)
```

```python
print(nivel_estudios)
```

```python
# DIAGRAMA DE BARRAS
import matplotlib.pyplot as plt
import numpy as np
estudios, frecuencia = np.unique(nivel_estudios, return_counts=True)
print(estudios, frecuencia)
```

```python
plt.ylabel('Frecuencia absoluta')
plt.xlabel('Nivel de estudios')
plt.title('Diagrama de barras')
plt.bar(estudios, frecuencia) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.show()
```

```python
# DIAGRAMA DE SECTORES
colores = {'red', 'blue', 'orange', 'green', 'cyan'}
plt.pie(frecuencia, labels=estudios, colors=colores, autopct='%1.2f%%')
plt.title('Diagrama de sectores')
plt.show()
```

```python
import pandas as pd
```

```python
# 1. Transformamos el arreglo mezclado a un DataFrame de Pandas
# 'nivel_estudios' es el arreglo de NumPy que ya tienes desordenado en la celda anterior.
df = pd.DataFrame(nivel_estudios, columns=['Nivel_Estudios'])
```

```python
print(" 1. VISTA PREVIA DE LOS DATOS (df.head)")
# Muestra los primeros 5 registros del DataFrame para verificar la estructura
print(df.head())
```

```python
print(" 2. DIMENSIONES DEL DATAFRAME (df.shape) ")
# Devuelve una tupla (filas, columnas)
print(f"El DataFrame tiene {df.shape[0]} filas (registros) y {df.shape[1]} columna.")
```

```python
print("3. INFORMACIÓN ESTRUCTURAL Y TIPOS DE DATOS (df.info)")
# Muestra si hay valores nulos, el tipo de dato (object/texto) y uso de memoria
print(df.info())
```

```python
print("4. RESUMEN ESTADÍSTICO DE LA VARIABLE CUALITATIVA (df.describe)")
# Al ser texto, describe() muestra: conteo total, valores únicos, el valor más repetido (top) y su frecuencia
print(df.describe())
```

```python
print("5. TABLA DE FRECUENCIAS ABSOLUTAS (df.value_counts)")
# Cuenta cuántas veces aparece cada categoría y las ordenamos de mayor a menor
frecuencias_abs = df['Nivel_Estudios'].value_counts()
print(frecuencias_abs)
```

```python
print("6. TABLA DE FRECUENCIAS RELATIVAS / PROPORCIONES")
# Con 'normalize=True' obtenemos el porcentaje de cada categoría en el DataFrame
frecuencias_rel = df['Nivel_Estudios'].value_counts(normalize=True) * 100
print(frecuencias_rel.map('{:.2f}%'.format))
```

```python
df = pd.DataFrame(nivel_estudios, columns=['Nivel_Estudios'])

# EXTRAER CARACTERÍSTICAS RESUMIDAS EN TABLAS

# Creamos una tabla de frecuencias completa (Absoluta y Relativa %)
tabla_caracteristicas = pd.DataFrame({
    'Frecuencia Absoluta': df['Nivel_Estudios'].value_counts(),
    'Frecuencia Relativa (%)': (df['Nivel_Estudios'].value_counts(normalize=True) * 100).round(2)
})

# Asignamos un nombre al índice de la tabla para que quede perfectamente identificado
tabla_caracteristicas.index.name = 'Nivel de Estudios'
```

```python
print("VISTA PREVIA DEL DATASET ORIGINAL (Primeras 10 filas):")
display(df.head(10))

print("\nTABLA DE CARACTERÍSTICAS EXTRAÍDAS:")
display(tabla_caracteristicas)
```
